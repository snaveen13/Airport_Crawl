## Create an empty JSON Template for Airport Data

```python
## create empty dictionary
airport_json = {}

## create main headers
list_of_headers = ['Airport_Stats', 
                   'Flight_Stats', 
                   'Airline_Stats', 
                   'Airport_Facilities', 
                   'Visa_Information']
    
## Detailed stats features
airport_stats_list = ['Total_Runways', 
                      'Yearly_Delay_Percent', 
                      'Airline_Terminal']

flight_stats_list = ['Flight_Delay_Percent',
                     'Arrival_Departure_Stats']

airline_stats_list = ['Monthly_Flight_Delay_Percent',
                      'Monthly_Flight_Cancellation_Percent']

visa_information_list = ['Country_Specific_Visa']

airport_facilities_list = ['Car_Rental', 
                           'Public_Transport', 
                           'Lounge', 
                           'Duty_Free', 
                           'Total_Transit_Hotel', 
                           'Attractions']

list_of_items = [airport_stats_list,
                 flight_stats_list, 
                 airline_stats_list, 
                 airport_facilities_list, 
                 visa_information_list]



for header, item_list in zip(list_of_headers, list_of_items):
    airport_json[header] = {}
    for item in item_list:
        airport_json[header][item] = {}
        
```

```python
print(json.dumps(airport_json, indent=2))
```

1. It looks like chennai airport doesnt have terminal information. Looking for other ways to get the data.

```python
def runways(soup):

    para = soup.find_all('p')
    
    for i in para:
        if 'runways' in str(i):
            total_runway = (str(i)[str(i).find('runways')-3:str(i).find('runways')]).strip()
            return(total_runway)
```

```python
runways(soup=soup)
```
